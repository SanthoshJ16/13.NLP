```python
import pandas as pd
import numpy as np
import itertools
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df=pd.read_csv('news.csv', index_col=None)
```

```python
dataset=df.drop("Unnamed: 0",axis=1)
```

```python
dataset
```

```python
# Step 1: Define target
y=dataset["label"]
```

#### Splitting the Test and Train data set

```python
# Step 2: Train-Test Split (your code)
X_train, X_test, y_train, y_test = train_test_split(dataset['text'], y, test_size=0.33, random_state=53)
```

#### Train & Precdict NLP using TF-IDF with MultinomialNB

```python
# Step 3: TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)

# Step 4: Fit on training data & transform
tfidf_train = tfidf_vectorizer.fit_transform(X_train)

# Step 5: Transform test data (NO fit here)
tfidf_test = tfidf_vectorizer.transform(X_test)

# Step 6: Model creation
clf = MultinomialNB()

# Step 7: Train model
clf.fit(tfidf_train, y_train)

# Step 8: Predict
pred = clf.predict(tfidf_test)

# Step 9: Accuracy
score = metrics.accuracy_score(y_test, pred)
print("Accuracy: %0.3f" % score)

# Step 10: Confusion Matrix
cm = metrics.confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
print("Confusion Matrix:\n", cm)

# Step 11: Classification Report
print("Classification Report:\n")
print(classification_report(y_test, pred))

```
