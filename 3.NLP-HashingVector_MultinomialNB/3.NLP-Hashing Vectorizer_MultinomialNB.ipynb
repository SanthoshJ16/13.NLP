```python
import pandas as pd
import numpy as np
import itertools
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score,from sklearn import metrics
```

```python
df=pd.read_csv('news.csv', index_col=None)
```

```python
dataset=df.drop("Unnamed: 0",axis=1)
```

```python
dataset
```

```python
# Step 1: Define target
y=dataset["label"]
```

#### Splitting the Test and Train data set

```python
# Step 2: Train-Test Split (your code)
X_train, X_test, y_train, y_test = train_test_split(dataset['text'], y, test_size=0.33, random_state=53)
```

#### Train & Precdict NLP using Hashing Vectorizer with MultinomialNB

```python
# Step 3: Hashing Vectorizer
hash_vectorizer = HashingVectorizer(
    stop_words='english',
    alternate_sign=False   # ✅ Important for Naive Bayes
)

# Step 4: Transform data
# (No fit_transform needed)
hash_train = hash_vectorizer.transform(X_train)
hash_test = hash_vectorizer.transform(X_test)

# Step 5: Model
clf = MultinomialNB()

# Step 6: Train
clf.fit(hash_train, y_train)

# Step 7: Predict
pred = clf.predict(hash_test)

# Step 8: Accuracy
score = accuracy_score(y_test, pred)
print("Accuracy: %0.3f" % score)

# Step 9: Confusion Matrix
cm = confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
print("Confusion Matrix:\n", cm)

# Step 10: Classification Report
print("Classification Report:\n")
print(classification_report(y_test, pred))
```
